## 中值滤波


```python
import cv2
import numpy as np

girl=cv2.imread('./girl.jpg')

#中值滤波
new=cv2.medianBlur(girl,5)

cv2.imshow('show1',np.hstack((girl,new)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 双边滤波


```python
import cv2
import numpy as np

girl=cv2.imread('./girl.jpg')

#双边滤波
new=cv2.bilateralFilter(girl,7,20,50)

cv2.imshow('show1',np.hstack((girl,new)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
