## 卷积操作
图像的卷积其实就是用卷积核（其实就是一个设计出来的小矩阵）

步长：就是不完全卷积的时候会跳

卷积核一般都是奇数，否则会出现小数填充


```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

#这里我使用了一个边缘提取的卷积核
kernel=np.float32([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
new=cv2.filter2D(mountain,-1,kernel)

cv2.imshow('show1',np.hstack((mountain,new)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
