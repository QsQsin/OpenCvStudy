## 计算周长、面积


```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

gray=cv2.cvtColor(mountain,cv2.COLOR_BGR2GRAY)

#二值化，返回阈值和二值化后的图
thresh , binary=cv2.threshold(gray,150,255,cv2.THRESH_BINARY)

result,contours,hierarchy=cv2.findContours(binary , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)

#绘制轮廓会直接修改原图
copy=mountain.copy()
new=cv2.drawContours(copy,contours,-1,(0,0,255),2)

cv2.imshow('show1',np.hstack((mountain,new)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
