
## 高斯金字塔



```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

#分辨率减小的操作，向下采样
dst1=cv2.pyrDown(mountain)

#向上采样
dst2=cv2.pyrUp(mountain)

cv2.imshow('show1',mountain)
cv2.imshow('show2',dst1)
cv2.imshow('show3',dst2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 拉普拉斯金字塔




```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

#先缩小
dst1=cv2.pyrDown(mountain)

#再放大
dst2=cv2.pyrUp(dst1)
#原图减去高斯金字塔就是拉普拉斯金字塔
lap0=mountain-dst2
back=lap0+dst2
cv2.imshow('show1',mountain)
cv2.imshow('show2',lap0)
cv2.imshow('show3',back)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
