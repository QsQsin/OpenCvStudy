```python
import cv2
import numpy as np

dots=cv2.imread('./dots.jpg')

gray=cv2.cvtColor(dots,cv2.COLOR_BGR2GRAY)

#
thresh , binary=cv2.threshold(gray,150,255,cv2.THRESH_BINARY)

result,contours,hierarchy=cv2.findContours(binary , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)

#
copy=dots.copy()

cv2.drawContours(copy,contours,-1,(0,0,255),2)

##
approx=cv2.approxPolyDP(contours[2],20,closed=True)
#approx,
cv2.drawContours(copy,[approx],0,(255,0,0),2)

cv2.imshow('show1',copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
