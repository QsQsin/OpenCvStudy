## 图像全局二值化



```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

gray=cv2.cvtColor(mountain,cv2.COLOR_BGR2GRAY)

ret , dst =cv2.threshold(gray,80,255,cv2.THRESH_BINARY)

cv2.imshow('show1',np.hstack((gray,dst)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
