```python
import cv2
import numpy as np

#加载视频
video=cv2.VideoCapture('./cars.mp4')

#创建MOG对象
mog=cv2.bgsegm.createBackgroundSubtractorMOG()
#创建卷积核
kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
min_w=100
min_h=100
line_high=600
offset=6
cars=[]
sum=0

def center(x,y,w,h):
    x1=int(w/2)
    y1=int(h/2)
    cx=int(x)+x1
    cy=int(y)+y1
    return cx,cy
while True:
    ret , frame = video.read()
    if ret == True:
        #灰度化，去噪
        frame_gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        frame_gray_blur=cv2.GaussianBlur(frame_gray,(3,3),5)
         #去背景
        frame_gray_blur_mog=mog.apply(frame_gray_blur)
        #腐蚀
        erode=cv2.erode(frame_gray_blur_mog,kernel)
        #膨胀
        dialte=cv2.dilate(erode,kernel)
        #闭运算
        close=cv2.morphologyEx(dialte,cv2.MORPH_CLOSE,kernel)
        #查找轮廓
        result,contours,h=cv2.findContours(close,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
        #画出检测线
        cv2.line(frame,(0,line_high),(1400,line_high),(255,0,0),3)
        #画出轮廓
        for contour in contours:
            #最大外接矩形
            (x,y,w,h)=cv2.boundingRect(contour)
            #过滤小的
            is_valid=(w>min_w) & (h>min_h)
            if not is_valid:
                continue
            cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255),2)
            #得到矩形的中心点
            cpoint=center(x,y,w,h)
            cars.append(cpoint)
            cv2.circle(frame,(cpoint),5,(0,0,255),-1)
            for (x,y) in cars:
                if y>(line_high-offset) and y < (line_high+offset):
                    sum+=1
                    cars.remove((x,y))
                    print(sum)

        cv2.putText(frame,'Car Number:'+str(sum),(500,60),cv2.FONT_HERSHEY_SIMPLEX,2,(128,0,128),5)
        cv2.imshow('show1',frame)
    key=cv2.waitKey(10)
    if key == 27:
        break

video.release()
cv2.destroyAllWindows()
```
