```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')
#
new1=cv2.boxFilter(mountain,-1,(5,5),normalize=True)
#
new2=cv2.blur(mountain,(5,5))

cv2.imshow('show1',np.hstack((mountain,new1,new2)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

#
new1=cv2.GaussianBlur(mountain,(5,5),sigmaX=1)
new2=cv2.GaussianBlur(mountain,(5,5),sigmaX=1000)

cv2.imshow('show1',np.hstack((mountain,new1,new2)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
