## 开运算
开运算=腐蚀+膨胀
提供了另一种去除噪声的思路

```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

#可以先腐蚀再膨胀，也可以直接API
new=cv2.morphologyEx(mountian,cv2.MORPH_RECT,iterations=3)

cv2.imshow(new)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 闭运算
闭运算=膨胀+腐蚀
可以去除图形内部噪声

```python
import cv2
import numpy as np

dots=cv2.imread('./dots.jpg')

kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
#可以先腐蚀再膨胀，也可以直接API
new=cv2.morphologyEx(dots,cv2.MORPH_CLOSE,kernel,iterations=3)

cv2.imshow('show1',np.hstack((dots,new)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
