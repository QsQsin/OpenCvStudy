```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

gray=cv2.cvtColor(mountain,cv2.COLOR_BGR2GRAY)

ret , dst =cv2.threshold(gray,80,255,cv2.THRESH_BINARY)

cv2.imshow('show1',np.hstack((gray,dst)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

gray=cv2.cvtColor(mountain,cv2.COLOR_BGR2GRAY)

adap=cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,3,0)

cv2.imshow('show1',np.hstack((gray,adap)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
