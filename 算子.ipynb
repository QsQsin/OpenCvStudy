## sobel算子
通过采用离散差分算子近似图像梯度



```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')
#使用算子,sobel只能分开计算x,y的梯度
dx=cv2.Sobel(mountain,cv2.CV_64F,dx=1,dy=0,ksize=5)
dy=cv2.Sobel(mountain,cv2.CV_64F,dx=0,dy=1,ksize=5)

dst=cv2.add(dx,dy)

cv2.imshow('show1',mountain)
cv2.imshow('show2',np.hstack((dx,dy,dst)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Scharr算子


```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

#使用算子,进阶的sobel，也是只能分开计算x,y的梯度
dx=cv2.Scharr(mountain,cv2.CV_64F,dx=1,dy=0)
dy=cv2.Scharr(mountain,cv2.CV_64F,dx=0,dy=1)

dst=cv2.add(dx,dy)

cv2.imshow('show1',mountain)
cv2.imshow('show2',np.hstack((dx,dy,dst)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 拉普拉斯算子



```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

#使用算子,可以同时计算x,y的梯度
new=cv2.Laplacian(mountain,cv2.CV_64F)

cv2.imshow('show1',np.hstack((mountain,new)))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
