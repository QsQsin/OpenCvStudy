## 腐蚀操作


```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

#定义核
kernel= np.ones((3,3),np.uint8)

#用定义的核进行腐蚀,iterations就是次数
new=cv2.erode(mountain,kernel,iterations=1)

cv2.imshow('show1',np.hstack((mountain,new)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## 膨胀操作


```python
import cv2
import numpy as np

mountain=cv2.imread('./mountain.jpg')

#定义核
kernel= np.ones((3,3),np.uint8)

#用定义的核进行腐蚀,iterations就是次数
new=cv2.dilate(mountain,kernel,iterations=1)

cv2.imshow('show1',np.hstack((mountain,new)))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
